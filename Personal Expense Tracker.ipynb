```python
import csv
import matplotlib.pyplot as plt

expenses = []
budget = 0
```

```python

# -------- Add Expense --------
def add_expense():

    date = input("Enter date (YYYY-MM-DD): ")
    category = input("Enter category(Food, Travel, etc..): ")
    amount = float(input("Enter amount: "))
    description = input("Enter description: ")

    expense = {
        'date': date,
        'category': category,
        'amount': amount,
        'description': description
    }

    expenses.append(expense)

    print("Expense added successfully!")
```

```python
# -------- View Expenses --------
def view_expenses():

    if not expenses:
        print("No expenses recorded.")
        return

    print("\nExpense List")

    for i, exp in enumerate(expenses, start=1):

        print(i, exp['date'], exp['category'], exp['amount'], exp['description'])
        print("------------------------")
```

```python
# -------- Delete Expense --------
def delete_expense():

    if not expenses:
        print("No expenses to delete.")
        return

    view_expenses()

    try:
        num = int(input("Enter expense number to delete: "))

        if 1 <= num <= len(expenses):
            deleted = expenses.pop(num - 1)
            print("Deleted:", deleted['category'], deleted['amount'])
        else:
            print("Invalid number.")

    except ValueError:
        print("Enter a valid number.")
```

```python
# -------- Set Budget --------
def set_budget():

    global budget

    budget = float(input("Enter your monthly budget: "))

    print("Budget stored:", budget)
```

```python
# -------- Track Budget --------
def track_budget():

    if budget == 0:
        print("Please set your budget first.")
        return

    total = 0

    for exp in expenses:
        total += exp['amount']

    print("Total Expenses:", total)
    print("Budget:", budget)

    if total > budget:
        print("Warning: You have exceeded your budget!")
    else:
        remaining = budget - total
        print("You have", remaining, "left for the month")
```

```python
# -------- Show Expense Graph --------
def show_expense_graph():

    if not expenses:
        print("No expenses recorded.")
        return

    summary = {}

    for exp in expenses:
        category = exp['category']
        amount = exp['amount']

        if category in summary:
            summary[category] += amount
        else:
            summary[category] = amount

    categories = list(summary.keys())
    amounts = list(summary.values())

    plt.bar(categories, amounts)

    plt.title("Expense by Category")
    plt.xlabel("Category")
    plt.ylabel("Amount")

    plt.show()
```

```python
# -------- Clear All Expenses --------
def clear_expenses():

    global expenses

    confirm = input("Are you sure you want to delete all expenses? (yes/no): ")

    if confirm.lower() == "yes":
        expenses = []
        print("All expenses cleared.")
    else:
        print("Operation cancelled.")
```

```python
# -------- Save Expenses --------
def save_expenses():

    with open("expenses.csv", "w", newline="") as file:

        writer = csv.writer(file)

        writer.writerow(["date", "category", "amount", "description"])

        for exp in expenses:

            writer.writerow([
                exp['date'],
                exp['category'],
                exp['amount'],
                exp['description']
            ])

    print("Expenses saved successfully.")
```

```python
# -------- Load Expenses --------
def load_expenses():

    try:
        with open("expenses.csv", "r") as file:

            reader = csv.DictReader(file)

            for row in reader:

                expenses.append({
                    'date': row['date'],
                    'category': row['category'],
                    'amount': float(row['amount']),
                    'description': row['description']
                })

        print("Previous expenses loaded.")

    except FileNotFoundError:
        print("No previous expense file found.")
```

```python
# -------- Menu --------
def menu():

    while True:

        print("\n----- Personal Expense Tracker -----")
        print("1. Add Expense")
        print("2. View Expenses")
        print("3. Set Budget")
        print("4. Track Budget")
        print("5. Save Expenses")
        print("6. Show Expense Graph")
        print("7. Delete Expense")
        print("8. Clear All Expenses")
        print("9. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_expense()

        elif choice == "2":
            view_expenses()

        elif choice == "3":
            set_budget()

        elif choice == "4":
            track_budget()

        elif choice == "5":
            save_expenses()

        elif choice == "6":
            show_expense_graph()

        elif choice == "7":
            delete_expense()

        elif choice == "8":
            clear_expenses()

        elif choice == "9":
            save_expenses()
            print("Exiting program...")
            break

        else:
            print("Invalid choice. Try again.")
```

```python
# -------- Start Program --------
load_expenses()
menu()
```
